```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Import the image
img = cv2.imread('burano.jpg') #We created a variable named img and set it equal to the cv2.imread object
plt.imshow(img)  #This uses the matplotlib library to plot/display the img variable
```

```python
# Convert the image into gray scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #We created another "img_gray" variable that's set equal to the cv2 library object that uses the grayscaleColor function
plt.imshow(img_gray, cmap = 'pink') #cmap is a built-in variable that tells you the type of color. Cmap is basically colormap
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  #This changes the color spaces to "Red Green Blue"
plt.imshow(img_rgb)
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'pink')
plt.show()
```

```python
eczema_img = cv2.imread('hand_eczema.jpg')
eczema_img = cv2.cvtColor(eczema_img,cv2.COLOR_BGR2RGB) #We have to convert it because imread used BGR the reverse color space and the plot function ised RGB color spaces
plt.imshow(eczema_img)  #This uses the matplotlib library to plot/display the img variable
```

```python
eczema_img_gray = cv2.cvtColor(eczema_img,cv2.COLOR_BGR2GRAY) 
cv2.putText(eczema_img_gray,text ="Eczema",   org = (250, 250),
            fontFace = cv2.FONT_HERSHEY_DUPLEX, 
            fontScale = 2, 
            color = (0, 255, 0), 
            thickness = 2, 
            lineType = cv2.LINE_AA)
plt.imshow(eczema_img_gray,cmap = 'gray') #It's important to use cmap or COLOR Map because if you don't indicate the color space, yo u get an 80's neon-looking pic
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(eczema_img[:, :, i], cmap = 'gray')
    
plt.show()
```

```python
# Transform the image into HSV and HLS models
eczema_img_hsv = cv2.cvtColor(eczema_img, cv2.COLOR_BGR2HSV)
eczema_img_hls = cv2.cvtColor(eczema_img, cv2.COLOR_BGR2HLS)

# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(eczema_img_hsv)
ax2.imshow(eczema_img_hls)
plt.show()
```

```python
avg_color_per_row = np.average(eczema_img, axis=0)
avg_color = np.average(avg_color_per_row, axis=0)
print(avg_color)
plt.imshow(eczema_img)
#This is supposed to be the average color
```
